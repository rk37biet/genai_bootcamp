```python
import re
```

The pattern <.*?> is a regular expression (regex) used to match HTML tags. Here's a breakdown:

**<: Matches the opening angle bracket of an HTML tag**

**.*?: Matches any sequence of characters (. matches any character, * means zero or more occurences), but the ? makes it non-greedy. This ensures it matches the shortest possible sequence between < and >**

**>: Matches the closing angle bracket of an HTML tag**

***Example:
For the input `<p>Hello</p>`, the pattern `<.*?>` will match:
`<p>`
`</p>`
The function `remove_html_tags` `replaces these matches` with an `empty string`, effectively removing HTML tags from the text

```python
def remove_html_tags(text):
    pattern = re.compile('<.*?>')
    return pattern.sub("",text)
```

```python
remove_html_tags(text)
```

```python
data["review"]=data["review"].apply(remove_html_tags)
data["review"]
```

The pattern r'https?://\S+|www\.\S+' is a regular expression (regex) used to match URLs. Here's the explaimation:

https?://:

Matches the literal "http://" or "https://"
The s? makes the "s" optional, so it matches both "http" and "https"

\S+:
Matches one or more non-whitespace characters (\S means any character that is not a space)
Esnures the URL continues until a space or end of the text

|:
Acts as an OR operator. Matches either the first pattern (https?://\S+) or the second

www\.\S+:
Matches URLs starting with "www.", followed by one or more non-whitespace characters.
The \. matches the literal dot . after "www".

```python
def remove_url(text):
    pattern=re.compile(r'https?://\S+|www\.\S+')
    return pattern.sub("",text)
```

```python
remove_url(text2)
```

```python
chat_words = {
    "AFAIK": "As Far As I Know",
    "AFK": "Away From Keyboard",
    "ASAP": "As Soon As Possible",
    "BTW": "By The Way",
    "B4": "Before",
    "LAMO": "Laughing My A** Off",
    "FYI": "For Your Information"
}
```

```python
chat_words["ASAP"]
```

```python
def chat_conversion(text):
    new_text = []
    for w in text.split():
        if w.upper() in chat_words:
            new_text.append(chat_words[w.upper()])
        else:
            new_text.append(w)
    return " ".join(new_text)
```

```python
def remove_stopwords(text):
    new_text = []
    for word in text.split():
        if word in stopwords.words("english"):
            new_text.append("")
        else:
            new_text.append(word.strip())
    return " ".join(new_text).replace("  "," ").replace("  "," ")
```

### Stemming and Lemmitization

#### Stemming
Keeping the data into the root word with dictionary meaning

```python
input_sentence="The quick brown foxes are jumping over the lazi dogs"
```

```python
from nltk.stem import PorterStemmer

def stemming(text):
    obj=PorterStemmer()

    stem_word=[obj.stem(word) for word in text.split()]

    return stem_word 
```

```python
stemming(input_sentence)
```

#### Lemmitization
Keeping the word into base/root form maintaining dictionary meaning or context

```python
from nltk.stem import WordNetLemmatizer

def lemmatization(text):
    words=text.split()

    lemmatizer=WordNetLemmatizer()
    lemetized_word=[lemmatizer.lemmatize(word) for word in words]
    
    return lemetized_word
```

```python
lemmatization(input_sentence)
```
