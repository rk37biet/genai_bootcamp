| Data Type                  | Dtype           | Description                                                                 |
|----------------------------|-----------------|----------------------------------------------------------------------------|
| 32-bit Floating Point      | `torch.float32` | Standard floating-point type for deep learning. Balances precision and memory. |
| 64-bit Floating Point      | `torch.float64` | Double-precision floating point for high-precision tasks. Uses more memory. |
| 16-bit Floating Point      | `torch.float16` | Half-precision for mixed-precision training. Reduces memory on GPUs. |
| BFloat16                   | `torch.bfloat16`| Brain floating-point for mixed-precision, especially on TPUs. |
| 8-bit Floating Point       | `torch.float8`  | Ultra-low-precision for experimental, memory-constrained environments. |
| 8-bit Integer              | `torch.int8`    | Signed integer for quantized models. Saves memory in inference. |
| 16-bit Integer             | `torch.int16`   | Signed integer for special numerical tasks with intermediate precision. |
| 32-bit Integer             | `torch.int32`   | Standard signed integer for indexing and general tasks. |
| 64-bit Integer             | `torch.int64`   | Long integer for large indexing or large numbers. |
| 8-bit Unsigned Integer     | `torch.uint8`   | Unsigned integer for image data (0-255). |
| Boolean                    | `torch.bool`    | Stores True/False for masks in logical operations. |
| Complex 64                 | `torch.complex64` | Complex numbers with 32-bit parts for scientific tasks. |
| Complex 128                | `torch.complex128`| Complex numbers with 64-bit parts for higher precision. |
| Quantized Integer          | `torch.qint8`   | Quantized signed 8-bit for efficient inference. |
| Quantized Unsigned Integer | `torch.quint8`  | Quantized unsigned 8-bit for image-related tasks. |

````python
# row wise sum of all the values in tensor e
torch.sum(e, dim=1)
````

````python
# row wise mean of tensor e
torch.mean(e, dim=1)
````

````python
# column wise mean of tensor e
torch.mean(e, dim=0)
````

````python
h = torch.randint(size=(3,3), low=0, high=10, dtype=torch.float32)
````
