````python
import torch.nn as nn
````

````python
class Model(nn.Module):
    def __init__(self,new_features):
        super().__init__()
        # create input layer or hidden layer using Linear function
        # input layer with 5 neurons/features and hidden layer with 1 neuron/feature
        self.linear = nn.Linear(in_features=new_features, out_features=1)
        # define an activation function
        self.sigmoid = nn.Sigmoid()
    def forward(self, features):
        # features means input data
        out = self.linear(features)
        out = self.sigmoid(out)
        return out
````

````python
data=torch.rand(10,5)
````

````python
data.shape
````

````python
model=Model(new_features=data.shape[1])
````
