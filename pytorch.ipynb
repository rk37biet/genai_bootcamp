````python
import torch.nn as nn
````

````python
class Model(nn.Module):
    def __init__(self,new_features):
        super().__init__()
        # create input layer or hidden layer using Linear function
        # input layer with 5 neurons/features and hidden layer with 1 neuron/feature
        # hidden layer nodes/neurons equals to number of output features
        self.linear = nn.Linear(in_features=new_features, out_features=1)
        # define an activation function
        self.sigmoid = nn.Sigmoid()
    def forward(self, features):
        # features means input data
        out = self.linear(features)
        out = self.sigmoid(out)
        return out
````

### Other way of defining NN model using Sequnetial method

For deep, and complex NN its recommned to use Sequential & network methods to create model and perform forward propogation

````python
import torch
import torch.nn as nn

class Model(nn.Module):

    def __init__(self,num_features):

        super().__init__()
        self.network = nn.Sequential(
            nn.Linear(in_features=num_features,out_features=3),
            nn.ReLU(),
            nn.Linear(in_features=3, out_features=1),
            nn.Sigmoid()
        )

    def forward(self, features):
        out = self.network(features)
        return out
````

````python
scaler=StandardScaler()
X_train= scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

encoder = LabelEncoder()
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
````

````python
# convert numpy array into tensor array

X_train_tensor = torch.from_numpy(X_train.astype(np.float32))
X_test_tensor = torch.from_numpy(X_test.astype(np.float32))
y_train_tensor = torch.from_numpy(y_train.astype(np.float32))
y_test_tensor = torch.from_numpy(y_test.astype(np.float32))
````

````python
class MySimpleNN(nn.Module):
    def __init__(self, new_features):
        super().__init__()
        self.linear1 = nn.Linear(in_features=new_features, out_features=3)
        self.relu = nn.ReLU()
        self.linear2 = nn.Linear(in_features=3,out_features=1)
        self.sigmoid = nn.Sigmoid()
    def forward(self, features):
        out = self.linear1(features)
        out = self.relu(out)
        out = self.linear2(out)
        out = self.sigmoid(out)
        return out
````

````python
X_train.shape[1]
````

````python
learning_rate = 0.1
epochs=1000
````

````python
loss_function=nn.BCELoss()
````

````python
model = MySimpleNN(new_features=X_train.shape[1])

optimizer = torch.optim.SGD(model.parameters(), lr=learning_rate)
````

````python
for epoch in range(epochs):
    # forward pass
    y_pred = model(X_train_tensor)
    
    # loss calculate
    loss = loss_function(y_pred, y_train_tensor.view(-1,1))

    # clear gradients/reset gradients after every epoch- helps to calcuate the gradients
    optimizer.zero_grad()

    # back-propgration- calculate gradients on top of loss
    loss.backward()

    # parameters update
    optimizer.step()

    # print loss in each epoch
    print(f'Epoch: {epoch + 1}, Loss: {loss.item()}')
````

### DataLoader & Dataset

Dataset -> Data initialization

DataLoader -> Mini batches

Gardient Decent-> we pass entire data in training

Mini batch Gradient Decent-> Passing data in mini batches -> Memory efficient -> Better convergence happens

````python
from sklearn.datasets import make_classification
import torch
````

````python
# Step:1 Create a synthetic classification dataset using sklearn
X, y = make_classification(
    n_samples=10,       # Number of samples
    n_features=2,       # Number of features
    n_informative=2,    # Number of informative features
    n_redundant=0,      # Number of redundant features
    n_classes=2,        # Number of classes
    random_state=42     # For reproducibility
)
````

````python
X
````

````python
y
````

````python
# Convert the data into PyTorch tensors
X = torch.tensor(X, dtype=torch.float32)
y = torch.tensor(y, dtype=torch.float32)
````

````python
X
````

````python
y
````

````python
from torch.utils.data import Dataset, DataLoader
````

````python
class CustomDataset(Dataset):
    def __init__(self, features, labels):
        # feature->X
        self.features=features
        # lables->y
        self.labels=labels
    def __len__(self):
        return self.features.shape[0]

    def __getitem__(self,index):
        return self.features[index],self.labels[index]
````

````python
dataset=CustomDataset(X,y)
````

````python
dataloader = DataLoader(dataset,batch_size=2,shuffle=False)
````
