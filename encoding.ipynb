```python
import pandas as pd
```

```python
data=pd.DataFrame({"text":["people watch cricket","cricket watch cricket","people give comment","cricket give comment"],"output":[1,1,0,0]})
```

```python
data
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
BOW=CountVectorizer()
```

```python
data["text"]
```

```python
document_matrix=BOW.fit_transform(data["text"])
```

```python
document_matrix.toarray()
```

```python
document_matrix[1].toarray()
```

vocabulary                         {'people': 3, 'watch': 4, 'cricket': 1, 'give': 2, 'comment': 0}<br>
alphabetical vocabulary            [comment, cricket, give, people, watch]<br>
d2                                 cricket watch cricket<br>
d2 vector representation           [0,2,0,0,1]<br>
document_matrix[1].toarray()       array([[0, 2, 0, 0, 1]])<br>
Conclusion                         Both are same as per conceptual and practical understanding of BOW
