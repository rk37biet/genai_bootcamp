### Sentence Transformer

```python
from sentence_transformers import SentenceTransformer
```

```python
model=SentenceTransformer("all-mpnet-base-v2")
```

```python
sentences[0]
```

```python
# vector representation of the sentence "I ate dinner"
model.encode(sentences[0])
```

```python
model.encode(sentences[0]).shape
```

### Langchain Hugginface

```python
from langchain_huggingface import HuggingFaceEmbeddings
```

```python
embedding_model_name = "sentence-transformers/all-mpnet-base-v2"
```

```python
model_kwargs = {"device": "cpu"}
embeddings = HuggingFaceEmbeddings(model_name=embedding_model_name, model_kwargs=model_kwargs)
```

```python
sentences[0]
```

```python
embedding_vector = embeddings.embed_query(sentences[0])
```

```python
embedding_vector
```

```python
len(embedding_vector)
```

### One more approach with Transformer Huggingface Library

```python
import torch
from transformers import BertTokenizer, BertModel
import random
```

```python
tokenizer=BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
model=BertModel.from_pretrained('bert-base-uncased')
```

```python
text = "follow sunny for genai learning"
```

```python
encoding = tokenizer([text],padding=True,truncation=True,return_tensors="pt",add_special_tokens=True)
```

```python
# input_ids are the token ids for each word in the sentence "follow sunny for genai learning"
input_ids=encoding["input_ids"]
input_ids
```

```python
# attention mask is telling each and every word is required from sentence "follow sunny for genai learning" for creating embedding vector
attention_mask=encoding["attention_mask"]
attention_mask
```

```python
# code to generate word based embedding vector for the sentence "follow sunny for genai learning"

with torch.no_grad():
    outputs=model(input_ids=input_ids, attention_mask=attention_mask)
    word_embeddings=outputs.last_hidden_state
```

```python
word_embeddings
```

```python
word_embeddings.shape
```

```python
# Lets decode word embedding vector to word token using tokenizer

decode=tokenizer.decode(input_ids[0],skip_special_tokens=True)
```
